## Evaluar Modelo Entrenado

**Reasoning**:

## Detectar Displays de Medidores

### Subtask:
Aplicar el modelo YOLOv12 entrenado a nuevas imÃ¡genes (o secuencias de imÃ¡genes/fotogramas de video) para detectar e identificar los displays de los medidores de luz.


**Reasoning**:
The subtask requires applying the trained YOLOv12 model for object detection. This step will use the loaded model to predict detections on the test set images as specified in the instructions, saving the results.



````python
import os

# Define the source for prediction (e.g., the test set images)
# Ensure dataset_path is correctly defined from previous steps
# For demonstration, we'll use the 'test/images' subfolder within the dataset_path.
source_for_prediction = os.path.join(dataset_path, 'test', 'images')

print(f"Starting prediction on: {source_for_prediction}")

# Perform object detection on the specified source
# The results will be saved to 'runs/detect/predict/' by default.
results = model.predict(source=source_for_prediction, conf=0.25, iou=0.7, save=True)

print("Detection complete. Results are saved in the 'runs/detect/predict/' directory.")
````

**Reasoning**:
The detection process is complete, and the results have been saved. To fulfill the task's requirement of visualizing the detections, the next step is to display some of these annotated images from the output directory.



````python
import matplotlib.pyplot as plt
import cv2

# Path to the directory where detection results are saved
# This is typically '/content/runs/detect/predict/' as indicated in the previous output.
results_dir = '/content/runs/detect/predict/'

# Get a list of detected image files
# Filter for image files (e.g., .jpg, .png) in the results directory
image_files = [f for f in os.listdir(results_dir) if f.lower().endswith(('.png', '.jpg', '.jpeg'))]

if not image_files:
    print(f"No images found in {results_dir}. Please ensure detections were saved correctly.")
elif len(image_files) > 5:
    print(f"Found {len(image_files)} detected images. Displaying the first 5 for brevity.")
    images_to_display = image_files[:5]
else:
    print(f"Found {len(image_files)} detected images. Displaying all of them.")
    images_to_display = image_files

plt.figure(figsize=(15, 5 * len(images_to_display)))
for i, image_file in enumerate(images_to_display):
    img_path = os.path.join(results_dir, image_file)
    # OpenCV reads images in BGR format, matplotlib expects RGB
    img = cv2.imread(img_path)
    if img is not None:
        img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        plt.subplot(len(images_to_display), 1, i + 1)
        plt.imshow(img_rgb)
        plt.title(f"Detection on {image_file}")
        plt.axis('off')
    else:
        print(f"Could not read image: {image_file}")

plt.tight_layout()
plt.show()
print("Visualization of detected meter displays complete.")
````
